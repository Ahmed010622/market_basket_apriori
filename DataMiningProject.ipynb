```python
df.columns=['العناصر']
df
```

```python
transactions=[]
items=df['العناصر'].values
print(items)

for i in range (0,len(items)):
    transactions.append(items[i].split(","))   
print(transactions)
```

```python
from mlxtend.preprocessing import TransactionEncoder
encoder=TransactionEncoder()
tran=encoder.fit(transactions).transform(transactions)
print(tran)
```

```python
encodeddata=pd.DataFrame(data=tran,columns=encoder.columns_ ,dtype=int)
encodeddata.head(5)
```

```python
encodeddata
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
support=apriori(encodeddata,min_support=0.1,use_colnames=True)
support.sort_values(by='support',ascending=False)
```

```python
confidence=association_rules(support,metric='confidence',min_threshold=0.3)
confidence.sort_values(by='confidence',ascending=False)
```

```python
# confidence > 0.6 و lift > 1
strong_rules = confidence[(confidence['confidence'] > 0.6) & (confidence['lift'] > 1)]

#confidence <= 0.6 و lift <= 1
weak_rules = confidence[(confidence['confidence'] <= 0.6) & (confidence['lift'] <= 1)]

print("Strong Rules:")
print(strong_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(2))

print("\nWeak Rules:")
print(weak_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(2))
```

```python
import matplotlib.pyplot as plt

confidence.head(10).plot(kind='bar', x='antecedents', y='confidence')
plt.title("Top Rules by Confidence")
plt.show()
```
